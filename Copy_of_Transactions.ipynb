```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample data
data = {
    "Mode": ["Cash", "Cash", "Saving Bank account 1", "Saving Bank account 1", "Cash"],
    "Category": ["Transportation", "Food", "subscription", "subscription", "Festivals"],
    "Subcategory": ["Train", "snacks", "Netflix", "Mobile Service Provider", "Ganesh Pujan"],
    "Note": ["2 Place 5 to Place 0", "Idli medu Vada mix 2 plates", "1 month subscription", "Data booster pack", "Ganesh idol"],
    "Amount": [30.0, 60.0, 199.0, 19.0, 251.0],
    "Income/Expense": ["Expense", "Expense", "Expense", "Expense", "Expense"],
    "Currency": ["INR", "INR", "INR", "INR", "INR"]
}

# Create a DataFrame from the sample data
df = pd.DataFrame(data)

# Group data by 'Category' and calculate the total 'Amount' for each category
category_totals = df.groupby('Category')['Amount'].sum()

# Create a bar chart to visualize the data
plt.figure(figsize=(10, 6))
category_totals.plot(kind='bar', color='skyblue')
plt.title('Category vs. Amount')
plt.xlabel('Category')
plt.ylabel('Total Amount (INR)')
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()

```
