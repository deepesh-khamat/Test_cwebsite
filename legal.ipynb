```python
import matplotlib.pyplot as plt
from collections import Counter

# Data from the extracted CSV

data = [
    ["Male", "30–40 years", "Legal Aid Lawyer", "Yes", "No", "Lack of awareness", "Yes", "Very Effective", "Lack of funding", "Yes", "Yes"],
    ["Female", "41–50 years", "NGO Representative", "Yes", "Not sure", "Financial constraints", "No", "Effective", "Limited public awareness", "Yes", "Not sure"],
    ["Male", "Below 30 years", "Social Worker/Activist", "No", "Yes", "Complex legal procedures", "Yes", "Very Effective", "Hostility or resistance from authorities", "No", "Yes"],
    ["Prefer not to say", "41–50 years", "Legal Aid Beneficiary", "Yes", "No", "Discrimination or bias", "Not sure", "Ineffective", "Bureaucratic challenges", "Yes", "Yes"],
    ["Female", "30–40 years", "Legal Aid Lawyer", "Yes", "Yes", "Lack of awareness", "Yes", "Effective", "Lack of funding", "Not sure", "Yes"],
    ["Male", "Above 50 years", "Academic/Researcher", "No", "Not sure", "Financial constraints", "Yes", "Neutral", "Limited public awareness", "Yes", "Yes"],
    ["Female", "Below 30 years", "Social Worker/Activist", "No", "Yes", "Complex legal procedures", "Not sure", "Very Effective", "Bureaucratic challenges", "Yes", "Yes"],
    ["Male", "41–50 years", "Government Official", "Yes", "No", "Lack of awareness", "Yes", "Effective", "Lack of funding", "Yes", "Yes"],
    ["Female", "30–40 years", "NGO Representative", "Yes", "Yes", "Financial constraints", "Yes", "Very Effective", "Limited public awareness", "No", "Yes"],
    ["Male", "Above 50 years", "Legal Aid Lawyer", "No", "Not sure", "Complex legal procedures", "Not sure", "Effective", "Bureaucratic challenges", "Yes", "Not sure"],
]

# Function to create pie chart
def create_pie_chart(data, column_index, title):
    counter = Counter(row[column_index] for row in data)
    labels, sizes = zip(*counter.items())

    plt.figure(figsize=(6,6))
    plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=["#ff9999","#66b3ff","#99ff99","#ffcc99"])
    plt.title(title)
    plt.show()

# Generating pie charts
create_pie_chart(data, 0, "Gender Distribution")
create_pie_chart(data, 1, "Age Group Distribution")
create_pie_chart(data, 2, "Occupation Distribution")
create_pie_chart(data, 3, "Legal Aid Involvement")
create_pie_chart(data, 4, "Effectiveness of Legal Aid")
create_pie_chart(data, 5, "Challenges in Accessing Legal Aid")
create_pie_chart(data, 6, "Judicial Intervention Effectiveness")
create_pie_chart(data, 7, "NGO Role Effectiveness")
create_pie_chart(data, 8, "NGO Challenges")
create_pie_chart(data, 9, "Need for Legal Aid Reform")
create_pie_chart(data, 10, "Support for Expanding Legal Aid")
```
